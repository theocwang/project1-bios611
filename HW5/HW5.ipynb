```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import SpectralClustering
from sklearn.metrics import pairwise_distances
from plotnine import ggplot, geom_point, aes
import pandas as pd

heroPowerData = pd.read_csv("/Users/Theo/Documents/Biology/BIOS 611/HW5/datasets_38396_60978_charcters_stats.csv")
heroPowerDataFinal = heroPowerData[["Intelligence", "Strength", "Speed", "Durability", "Power", "Combat", "Total", "Alignment"]]
heroPowerDataFinal.dropna(subset=["Alignment"], inplace=True)
heroPowerDataFinal["Alignment"] = heroPowerDataFinal["Alignment"].replace(["good"], 1)
heroPowerDataFinal["Alignment"] = heroPowerDataFinal["Alignment"].replace(["bad"], 2)
heroPowerDataFinal["Alignment"] = heroPowerDataFinal["Alignment"].replace(["neutral"], 0)
print(heroPowerDataFinal)

projection = TSNE(n_components=2).fit_transform(heroPowerDataFinal)
projDF = pd.DataFrame(projection, columns=["X1","X2"])
projDF["Alignment"] = heroPowerDataFinal["Alignment"]
(ggplot(projDF, aes("X1","X2")) + geom_point(aes(color="Alignment")))
print(projDF)

projDF.to_csv('data.csv', index=False, header=True)
```
